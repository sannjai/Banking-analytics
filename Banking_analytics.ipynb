** Outier Detection **

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def outlier_detection(Column1):
    Q1 = df[Column1].quantile(0.25)
    Q3 = df[Column1].quantile(0.75)
    IQR = Q3 - Q1  # Interquartile Range

    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df[(df[Column1] < lower_bound) | (df[Column1] > upper_bound)]
    print("Outliers:")
    print(outliers)

    # Plotting a boxplot to visualize the outliers
    plt.figure(figsize=(8, 6))
    plt.boxplot(df[Column1])
    plt.title('Boxplot of Column1')
    plt.ylabel('Values')
    plt.show()



    sns.histplot(df[Column1], kde=True)
    plt.title("Distribution of Column1")
    plt.show()
```

```python
categorical_columns = ['Type_of_Loan', 'Occupation', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour', 'Credit_Score']

# Print unique categories for each categorical column
for col in categorical_columns:
    print(f"Unique values in {col}:")
    print(df[col].unique())
    print("\n")
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, MultiLabelBinarizer
```

```python
# MultiLabelBinarizer for 'Type_of_Loan'
df['Type_of_Loan'] = df['Type_of_Loan'].fillna('')


mlb = MultiLabelBinarizer()
df['Type_of_Loan'] = df['Type_of_Loan'].apply(lambda x: x if isinstance(x, list) else str(x).split(', '))

loan_encoded = mlb.fit_transform(df['Type_of_Loan'])
loan_encoded_df = pd.DataFrame(loan_encoded, columns=mlb.classes_)

# Label Encoding for 'Occupation'
label_encoder = LabelEncoder()
df['Occupation_Encoded'] = label_encoder.fit_transform(df['Occupation'])

# Ordinal Encoding for 'Credit_Mix' and 'Credit_Score'
ordinal_encoder = OrdinalEncoder(categories=[['Bad', 'Standard', 'Good','Poor']])
df[['Credit_Mix_Encoded']] = ordinal_encoder.fit_transform(df[['Credit_Mix']])
df[['Credit_Score_Encoded']] = ordinal_encoder.fit_transform(df[['Credit_Score']])

# Label Encoding for 'Payment_of_Min_Amount'
df['Payment_of_Min_Amount_Encoded'] = label_encoder.fit_transform(df['Payment_of_Min_Amount'])

# Label Encoding for 'Payment_Behaviour'
df['Payment_Behaviour_Encoded'] = label_encoder.fit_transform(df['Payment_Behaviour'])

# Combine the MultiLabelBinarizer dataframe with the original dataframe
df = pd.concat([df, loan_encoded_df], axis=1)

# Drop original columns if needed
df.drop(['Type_of_Loan', 'Occupation', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour', 'Credit_Score'], axis=1, inplace=True)

print(df.head())
```

```python
df.drop(['Name','Customer_ID','SSN'], axis=1, inplace=True)
```

```python
df.set_index('ID', inplace=True)
```

```python
df.head(5)
```

```python
from sklearn.cluster import KMeans

inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 10), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.grid()
plt.show()
```

```python
optimal_k = 3  # For example, let's say we choose 3 clusters

# Fit the K-Means model
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(df)

# Step 4: Analyze the clusters
cluster_summary = df.groupby('Cluster').mean()
print(cluster_summary)

# Step 5: Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual_Income', y='Outstanding_Debt', hue='Cluster', palette='viridis', s=100)
plt.title('Customer Segmentation by Annual Income and Outstanding Debt')
plt.xlabel('Annual Income')
plt.ylabel('Outstanding Debt')
plt.legend(title='Cluster')
plt.show()
```

```python
# Step 5: Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual_Income', y='Num_of_Delayed_Payment', hue='Cluster', palette='viridis', s=100)
plt.title('Customer Segmentation by Annual Income and Credit_Score')
plt.xlabel('Annual Income')
plt.ylabel('Outstanding Debt')
plt.legend(title='Cluster')
plt.show()
```
